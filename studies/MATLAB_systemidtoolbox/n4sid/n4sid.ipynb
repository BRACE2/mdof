```python
import scipy
import numpy as np
from matplotlib import pyplot as plt
```

```python
n_decimations = 16
n_modes = len(scipy.io.loadmat(f"decimation_1.mat")["periods"])
periods = np.zeros((n_modes, n_decimations))
fig, ax = plt.subplots(figsize=(8,3.5))
for i in range(n_decimations):
    periodsi = scipy.io.loadmat(f"decimation_{i+1}.mat")["periods"][:,0]
    ax.plot(periodsi, np.arange(1,n_modes+1), "o", alpha = 0.5, label=f"d = {i+1}")
    periods[:,i] = periodsi
ax.set_xlabel("Period (s)")
ax.set_ylabel("Mode")
ax.set_yticks(np.arange(1,n_modes+1))
ax.set_title("Periods identified using MATLAB n4sid for Hayward Bridge Berkeley Event")
ax.legend(bbox_to_anchor=(1.2, 1.05));
```

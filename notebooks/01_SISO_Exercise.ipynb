parameter | value
----------|-------
m  | mass

The oscillator is subject to the following forcing function:

$f(t) = e^{-at}\sin{bt}$

```python
a = 1
b = np.pi
nt = 100
dt = 0.1
time = np.arange(0,nt*dt,dt)
force = np.exp(-a*time)*np.sin(b*time)
plt.plot(time, force)
plt.title(f"Forcing over time ({a=:.2f}, {b=:.2f})")
```
